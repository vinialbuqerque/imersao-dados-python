```python
renomear_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df.rename(columns = renomear_colunas, inplace = True)
df.columns
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
tamanho_empresa = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0:'Presencial',
    50:'Hibrido',
    100:'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

- Utilizando o método de média e mediana para calcular um valor que possa substituir os espaços nulos
- Utilizando a biblioteca numpy para inserir valores nulos no método

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#calcula a mediana e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

- Substituindo espaços nulos dentro do contexto período de uma semana e clima
- Preenchendo células nulas com base no passado (ffill)
- Preenchendo células nulas com base no futuro (bfill)

```python
df_temperaturas = pd.DataFrame ({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

- Desta vez, substituindo espaços nulos dentro do contexto de cidades e nomes cadastrados

```python
df_cidades = pd.DataFrame ({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
display(df_cidades)
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')
plt.title('Salário médio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (USD)')
plt.show()
```

- Criação de um gráfico em estilo histograma para representação da distribuição dos salários anuais

```python
plt.figure(figsize = (8, 4))
sns.histplot(df_limpo['usd'], bins = 50, kde = True)
plt.title('Distribuição dos salário anuais')
plt.xlabel('Salário (USD)')
plt.ylabel('Frequência')
plt.show()
```

- Criação de um gráfico no estilo boxplot
- Em seguida, refinando o mesmo para uma melhor visualização e interpretação da distribuição dos salários por senioridade

```python
plt.figure(figsize = (8, 5))
sns.boxplot(x = df_limpo['usd'])
plt.title('Boxplot salário')
plt.xlabel('Salário (USD)')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

plt.figure(figsize = (8, 5))
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade, palette = 'Set2', hue = 'senioridade')
plt.title('Distribuição salarial por nível de senioridade')
plt.xlabel('Salário (USD)')
plt.show()
```

- Criação de um gráfico interativo representando a média salarial anual por senioridade utilizando a biblioteca plotly

```python
import plotly.express as px
```

```python
senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending = False).reset_index()

fig = px.bar(senioridade_media_salario, x = 'senioridade', y = 'usd', title = 'Média Salarial por Senioridade', labels = {'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual (USD)'})
fig.show()
```

- Criação de um gráfico estilo pie para representação das proporções entre os tipos de trabalho da empresa
- Em seguida, a refinação da mesmo para uma melhor interpretação e visualização dos dados

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho'

          )

fig.show()
```

```python

remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho',
             hole=0.5
          )
fig.update_traces(textinfo='percent+label')
fig.show()
```
